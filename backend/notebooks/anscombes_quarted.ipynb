```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Anscombe_quartet_data.csv")
df
```

```python
def myplot(x,y):
    plt.ylim(0, 14)
    plt.xlim(0, 20)

    coef = np.polyfit(x,y,1)
    poly1d_fn = np.poly1d(coef) 
    
    plt.plot([0,20], poly1d_fn([0,20]), 'b')
    plt.plot(x, y, 'ko')

    ax = plt.gca()
    ax.axes.xaxis.set_ticklabels([])
    ax.axes.yaxis.set_ticklabels([])

    
```

```python
myplot(df["x123"],df["y1"])
plt.savefig("anscombes0.pdf", bbox_inches = "tight")
plt.show()
```

```python
myplot(df["x123"],df["y2"])
plt.savefig("anscombes1.pdf", bbox_inches = "tight")
plt.show()
```
