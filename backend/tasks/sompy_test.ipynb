```python
from sklearn import tree
```

```python
X = [[0,0], [1,1]]
Y = [0,1]
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
clf = clf.fit(X,Y)
```

```python
clf.predict([[2.,2.]])
```

```python
tree.plot_tree(clf)
```
